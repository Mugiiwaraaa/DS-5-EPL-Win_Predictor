## Building a model that predicts the chances of winning a match for your favourite EPL team.
In my case, it's Manchester City

```python
#Importing the dependancies 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
```

```python
city = pd.read_csv('Man_City_AllSeason.csv')
```

### Some basic data analysis and preprocessing 

```python
#making sure there are no null values in our dataset 
city.isnull().sum()
```

```python
city.info()
```

```python
#dropping some unnecessary columns that Ii dont require 
# Using the date of the match as my index 
data = city.drop(columns=['Div','Date','Referee'])
```

```python
data
```

```python
data.tail(5)
```

## Calculating Man city's total Wins,Losses and Draws in last 12 seasons

```python
def matchDecision(row):
    value = row[1]
    if value[0] == 'Man City':
        if value[2] == 'H':
            return 'W'
        elif value[2] == 'A':
            return 'L'
        else:
            return 'D'
    
    if value[1] == 'Man City':
        if value[2] == 'H':
            return 'L'
        elif value[2] == 'A':
            return 'W'
        else:
            return 'D'
        
wld = []
for row in data[['HomeTeam','AwayTeam','FTR']].iterrows():
    wld.append(matchDecision(row))   
```

```python
data['WLD'] = wld
```

```python
data['WLD']
```

```python
data
```

## Manipulating Dataset to be represented as integer datatype as follows:
1. Win : 1
2. Loss : -1
3. Draw : 0
4. If ManCity is HomeTeam-> 1
5. If Mancity is AwayTeam-> 0

```python
label ={"W": +1.0,"L":-1.0,"D":0.0}
data['Label']=data['WLD'].map(label)
data.drop(columns=['WLD'],inplace=True)
```

```python
data['Home'] = data['HomeTeam'] == 'Man City'
data["Home"] = data["Home"].astype(int)
```

```python
data['Away'] = data['AwayTeam'] == 'Man City'
data["Away"] = data["Away"].astype(int)
```

```python
data.drop(columns=['HomeTeam','AwayTeam','FTR','HTR'],inplace=True)
```

# Splitting the dataset into training and testing data 


```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(data,test_size=0.15,random_state=42)
print(f"rows in test_set:{len(train_set)}\nrows in test_Set:{len(test_set)}")
```

# Stratified shuffling 
This is to make sure there are similar ratios in both test and train sets

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.15,random_state=42)
for train_index ,test_index in split.split(data,data['Label']):
    strat_train_set=data.loc[train_index]
    strat_test_set=data.loc[test_index]
```

```python
strat_test_set['Label'].value_counts()
```

```python
strat_train_set['Label'].value_counts()
```

Therefore we get similar ratios of wins,losses and draws for both train and test sets.

## Splitting the Features and labels of our dataset 

```python
train_data_features=strat_train_set.drop('Label' , axis=1)
train_data_label = strat_train_set['Label'].copy()
test_data_features=strat_test_set.drop('Label' , axis=1)
test_data_label = strat_test_set['Label'].copy()
```

# Creating a Win Predictor 

```python
train_data_label_W= (train_data_label==1.0)
test_data_label_W=(test_data_label==1.0)
```

## Creating a pipeline to automate some processes

```python
## Automating processes and feature scaling
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([
    ('std_scaler',StandardScaler())
])
```

```python
train_data_features_tr=my_pipeline.fit_transform(train_data_features)
```

```python
train_data_features_tr.shape
```

```python
train_data_features.shape
```

## Selecting and training a desired model for our predictor 

```python
#First I try it out with logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
# model = LogisticRegression()
model = GaussianNB()
# model = RandomForestClassifier()
# model = svm.SVC()
# model = DecisionTreeClassifier()
model.fit(train_data_features_tr,train_data_label_W)
```

```python
#Testing the model on some training data
some_data=train_data_features.iloc[:5]
some_labels=train_data_label_W.iloc[:5]

```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
some_labels
```

```python
#Testing the model on some test data
some_testdata=test_data_features.iloc[:5]
some_testlabels=test_data_label_W.iloc[:5]
prepared_testdata=my_pipeline.transform(some_testdata)
```

```python
model.predict(prepared_testdata)
```

```python
some_testlabels
```

### 3 out of 5 outcomes were predicted right, which is decent 

```python
from sklearn.model_selection import cross_val_score
a=cross_val_score(model,train_data_features,train_data_label_W,cv=3,scoring='accuracy')
```

```python
a.mean()
#This is the acuracy of the model.
```

#### As we know for a classifier, 'accuracy' is not the best metric to evaluate it. Therefore, We calculate the F1-Score 

```python
from sklearn.model_selection import cross_val_predict
train_data_pred=cross_val_predict(model,train_data_features,train_data_label_W,cv=3)
train_data_pred
```

## Calculating the confusion matrix

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
confusion_matrix(train_data_label_W,train_data_pred)
disp = ConfusionMatrixDisplay.from_predictions(train_data_label_W,train_data_pred)
disp.plot()
plt.xlabel=("Predicted")
plt.ylabel=("Actual")
plt.show()
```

## Calculating Precision , Recall and F1-Scores

```python
from sklearn.metrics import precision_score , recall_score , f1_score
#Calculating precision score
precision_score(train_data_label_W,train_data_pred)
```

```python
#Calculating recall score
recall_score(train_data_label_W,train_data_pred)
```

```python
#Calculating the F-1 scores
f1_score(train_data_label_W,train_data_pred)
```

```python
from sklearn.metrics import precision_score , recall_score , f1_score
#Calculating precision score
precision_score(test_data_label_W,test_data_pred)
```

```python
#Calculating recall score
recall_score(test_data_label_W,test_data_pred)
```
